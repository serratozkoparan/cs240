```python
import thinkplot
```

```python
import thinkstats2
```

```python
width = 0.5
hist = thinkstats2.Hist(resp.fmarno, label='The number of marriages')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='# of marriages', ylabel='Count of respondents')
```

```python
neverMarried = resp[resp.fmarno == 0]
```

```python
married = resp[resp.fmarno >= 1]
```

```python
first_hist = thinkstats2.Hist(neverMarried.totincr, label='neverMarried')
other_hist = thinkstats2.Hist(married.totincr, label='married')
width = 0.5
thinkplot.PrePlot(2)
thinkplot.Hist(first_hist, align='right', width=width)
thinkplot.Hist(other_hist, align='left', width=width)
thinkplot.Config(xlabel='# of Respondents', ylabel='Total Income')
```

```python
print "The min, max, std. dev, variance and mean of total income of neverMarried respondents are %d,%d,%d,%d,%d." %(neverMarried.totincr.min(),neverMarried.totincr.max(),neverMarried.totincr.std(),neverMarried.totincr.var(),neverMarried.totincr.mean())
```

```python
print "The min, max, std. dev, variance and mean of total income of married respondents are %d,%d,%d,%d,%d." %(married.totincr.min(),married.totincr.max(),married.totincr.std(),married.totincr.var(),married.totincr.mean())
```
