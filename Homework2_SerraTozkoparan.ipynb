```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first

import thinkstats2
import thinkplot
```

```python
resp = nsfg.ReadFemResp()
```

```python
school = resp[resp.hieduc <= 10]
university = resp[resp.hieduc > 10]
```

```python
school.agemomb1.min(), school.agemomb1.max(), school.agemomb1.mean(), school.agemomb1.median(), school.agemomb1.var() 
```

```python
university.agemomb1.min(), university.agemomb1.max(), university.agemomb1.mean(), university.agemomb1.median(), university.agemomb1.var()
```

```python
first_pmf = thinkstats2.Pmf(school.agemomb1, label='school')
other_pmf = thinkstats2.Pmf(university.agemomb1, label='university')
width = 0.5
thinkplot.PrePlot(2)
thinkplot.Pmfs([first_pmf, other_pmf])
thinkplot.Config(xlabel='ages of mothers at first birth', ylabel='PMF')
```

```python
first_cdf = thinkstats2.Cdf(school.agemomb1, label='school')
second_cdf = thinkstats2.Cdf(university.agemomb1, label='university')
width = 0.5
thinkplot.PrePlot(2)
thinkplot.Cdfs([first_cdf, second_cdf]) 
thinkplot.Config(xlabel='ages of mothers at first birth', ylabel='CDF')
```

```python
cdf = thinkstats2.Cdf(resp.age_a, label='age')
thinkplot.Cdf(cdf, complement=True)
thinkplot.Config(xlabel='Age of the woman being interviewed',
                 ylabel='CCDF', yscale='log', loc='upper right')

#It is consistent with exponential distribution. 
```

```python
cdf = thinkstats2.Cdf(resp.basewgt, label='data')

thinkplot.PrePlot(1)
thinkplot.Cdf(cdf) 
thinkplot.Config(title='Weights',
                 xlabel='Weight of the woman being interviewed',
                 ylabel='CDF')

#It is consistent with a pareto distribution.
```
